### Working with sintetic seimic data

```python
def visualize_3d_array(data):

    # Create a figure and a 3D Axes
    fig = plt.figure()
    # fig = plt.figure(projection='3d')
    ax = fig.add_subplot(111, projection='3d')

    # Define the x, y, and z coordinates
    x, y, z = np.meshgrid(np.arange(data.shape[0]),
                          np.arange(data.shape[1]),
                          np.arange(data.shape[2]))

    # Plot the volume
    ax.scatter(x, y, z, c=data.flatten(), cmap='rainbow')
    # Show the plot
    plt.show()
```

```python
model_qube = np.zeros((100, 50, 10))
```

```python
visualize_3d_array(model_qube)
```

```python
model_qube[:, :, 2:6] = 1
```

```python
visualize_3d_array(model_qube)
```

```python
model_qube[:, 30:35, :] = 2
```

```python
visualize_3d_array(model_qube)
```

## Exploting dimensions

```python
model_qube = np.zeros((100, 50, 10), dtype=int)
```

```python
# Create a 3D NumPy array

# Iterate through the dimensions using nditer
it = np.nditer(model_qube, flags=['multi_index'], op_flags=['readwrite'])

value_counter = 0
while not it.finished:
    
    value_counter += 1
    # Access the value and its index
    value = it[0]
    indices = it.multi_index
    
    # Modify the value (for example, double it)
    it[0] = value_counter
    
    # Move to the next element
    it.iternext()
```

```python
visualize_3d_array(model_qube[:, :, :])
```

```python
visualize_3d_array(model_qube[0:1, :, :])
```

## Calculating reflection coefficient

```python
rpp_matrix = np.zeros(matrix.shape)

for row in range(matrix.shape[0]):
    for column in range(matrix.shape[1]):
        for layer in range(matrix.shape[2] - 1):
            rpp_matrix[row, column, layer] = matrix[row, column, layer + 1] - matrix[row, column, layer]
```

```python
visualize_3d_array(rpp_matrix)
```

```python
plt.figure(figsize=(15, 15), dpi=300)
plt.subplot(121)
plt.imshow(rpp_matrix[0, :, :], cmap="rainbow")

plt.subplot(122)
plt.imshow(rpp_matrix[:, -1, :], cmap="rainbow")
```
